__Q1__
(From exercises 9.5.2 of Advanced Modern Engineering Mathematics)

a) Show that the function $\phi(x,y)=x^2y$ satisfied the 2-d Poisson equation

\begin{align}
\nabla^2 \phi = \frac{\partial^2 \phi(x,y)}{\partial x^2} + \frac{\partial^2 \phi(x,y)}{\partial y^2} = 2y & & (1)
\end{align}

__Solution__

First, find the first and second partial differentials of $\phi(x,y)$

\begin{align}
\frac{\partial \phi(x,y)}{\partial x} &= 2xy & \frac{\partial^2 \phi(x,y)}{\partial x^2} &= 2y & \\
\frac{\partial \phi(x,y)}{\partial y} &= x^2 & \frac{\partial^2 \phi(x,y)}{\partial y^2} &= 0 & 
\end{align}

Then substitute these into (1) to find

\begin{align}
\frac{\partial^2 \phi(x,y)}{\partial x^2} + \frac{\partial^2 \phi(x,y)}{\partial y^2} = 2y
\end{align}

thus showing the function $\phi(x,y)=x^2y$ satisfies the 2-d Poission equation (1).

b) By putting $\phi(x,y) = u(x,y) + x^2y$, show that $u(x,y)$ satisfies the Laplace equation. Then use the method of separated solutions to find the solution for $\phi(x,y)$ in the unit square which satisfies the boundary conditions

\begin{align}
\phi(0, y) &= 0, & 0 \le x \le 1 && (2) \\
\phi(1, y) &= y, & 0 \le x \le 1 && (3) \\
\phi(x, 0) &= 0, & 0 \le y \le 1 && (4) \\
\phi(x, 1) &= x^2 + \sin\pi x, & 0 \le y \le 1 && (5)
\end{align}
\
\
__Solution__

If $\phi(x,y)=u(x,y)+x^2y$, then since $\phi(x,y)=x^2y$ satisfies the inhomogeneity $2y$, it must be the case that $u(x,y)$ satisfies the Laplace equation (i.e. without any inhomogeneity). To prove this, set $u(x,y)=\phi(x,y)-x^2y$ and then find the first and second partial derivatives

\begin{align}
\frac{\partial u(x,y)}{\partial x} &= \frac{\partial \phi(x,y)}{\partial x}-2xy & \frac{\partial^2 u(x,y)}{\partial x^2} &= \frac{\partial^2 \phi(x,y)}{\partial x^2}-2y & \\
\frac{\partial u(x,y)}{\partial y} &= \frac{\partial \phi(x,y)}{\partial y}- x^2 & \frac{\partial^2 u(x,y)}{\partial y^2} &= \frac{\partial^2 \phi(x,y)}{\partial y^2} & 
\end{align}

Thus

\begin{align}
\frac{\partial^2 u(x,y)}{\partial x^2} + \frac{\partial^2 u(x,y)}{\partial y^2} = \frac{\partial^2 \phi(x,y)}{\partial x^2} + \frac{\partial^2 \phi(x,y)}{\partial y^2}-2y &= 2y - 2y =0
\end{align}

which proves that $u(x,y)$ satisfies the Laplace equation as required.

To solve the Laplace equation for $u(x,y)$ we must first transform the boundary conditions (2)-(5)

\begin{align}
u(0, y) &= \phi(0,y) - 0 = 0-0 = 0, & 0 \le x \le 1 && (6) \\
u(1, y) &= \phi(1,y) - y = y-y = 0, & 0 \le x \le 1 && (7) \\
u(x, 0) &= \phi(x,0) - 0 = 0-0 = 0, & 0 \le y \le 1 && (8) \\
u(x, 1) &= \phi(x,1) - x^2 = x^2 + \sin\pi x - x^2 = \sin\pi x, & 0 \le y \le 1 && (9)
\end{align}

Now we can solve the Laplace equation for $u(x,y)$. From the lectures or the book we know that the following three functions

\begin{align}
u_1(x,y) =& (A\sin \mu x+B\cos \mu x) \left(Ce^{\mu y}+De^{-\mu y}\right) \\
u_2(x,y) =&\left(Ae^{\mu x}+Be^{-\mu x}\right)(C\sin \mu y+D\cos \mu y) \\
u_3(x,y) =& (Ax+B)(Cy+D)
\end{align}

are possible solutions for the Laplace equation.

We need to go through the boundary conditions to find which of these three functions is a solution to the partial differential equation with associated boundary conditions. In addition, we need to use the boundary conditions to find the values for the parameters, i.e. $A$, $B$, $C$, $D$ and $\mu$, in these functions.

The first two boundary conditions (6) and (7) require that the solution has $\sin$ and $\cos$ terms in $x$ since these terms have multiple zeros. Conversely, $\sinh$ and $\cosh$ have only $1$ or $0$ roots and are thus not suitable for the term in $x$ with these boundary conditions. Also $u_3$ is not suitable because it can only give zeros at $x=0$ and $x=1$ for all $y$ if $u_3(x,y)=0$, which contradicts boundary condition (9). Thus the solution must have the form,

$$
u_1(x,y) = (A\sin \mu x+B\cos \mu x) \left(Ce^{\mu y}+De^{-\mu y}\right)
$$

Dropping the subscript and using boundary condition (6) we get

\begin{align}
u(0,y) &= (A\sin(\mu 0)+B\cos(\mu 0)) \left(Ce^{\mu y}+De^{-\mu y}\right) \\
 &= B \left(Ce^{\mu y}+De^{-\mu y}\right) \overset{!}{=} 0 
\end{align}

This needs to hold for all $y\ge0$ which is only possible for $B=0$. We can thus write the solution as

$$
u(x,y) = \sin(\mu x) \left(\tilde{C}e^{\mu y} + \tilde{D}e^{-\mu y}\right)
$$

where $\tilde{C} = AC$ and $\tilde{D} = AD$.

Using boundary condition (7) we get,

$$
u(1,y) = \sin(\mu) \left(\tilde{C}e^{\mu y} + \tilde{D}e^{-\mu y}\right) \overset{!}{=} 0 
$$

Again this needs to hold for all $y\ge0$ which can only be fulfilled if $\sin(\mu)=0$. We know that the sine function is zero for multiples of $\pi$ and thus that $\mu=n\pi$ where $n=1,2,\dots$.

Using boundary condition (8) we get

$$
u(x,0) = \sin(n\pi x) \left(\tilde{C}e^{\mu 0} + \tilde{D}e^{-\mu 0}\right) \overset{!}{=} 0 
$$

which holds for $0\le x \le 1$ only if we have $\tilde{C} = -\tilde{D}$. Thus the solution has the form

\begin{align}
u(x,y) = 2\tilde{C}\sin(n\pi x) \sinh{n\pi y}, \quad n=1,2,\dots & & (10)
\end{align}

To satisfy boundary condition (9)

$$
u(x,1) = 2\tilde{C}\sin{n\pi x} \sinh{n\pi} = \sin{\pi x}  \quad 0\le x \le1
$$

we must have $n=1$ and $\tilde{C}=\frac{1}{2\sinh{\pi}}$

The complete solution is

$$
\phi(x,y) = \frac{\sin{\pi x} \sinh{\pi y}}{\sinh{\pi}} + x^2y
$$


c) Use Python to plot the solution.

__Solution__

The Python script is the same as for Exercise sheet 1 with the exception of a different function definition and different solution domain.

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator , FormatStrFormatter
import numpy as np
# Solution to the Poisson equation on the unit square
# Set the x and y grid points and the meshgrid
N = 100
x = np.linspace(0,1,N)
y = np.linspace(0,1,N)
X,Y = np.meshgrid(x,y)
# Define the function and generate the output on the meshgrid
Z = np.sin(np.pi * X) * np.sinh(np.pi * Y) / np.sinh(np.pi) + X*X*Y
# Generate the plot
fig = plt.figure()
ax = plt.axes(projection='3d')
# Plot the surface
surf = ax.plot_surface(X,Y,Z,cmap=cm.coolwarm,linewidth=0,antialiased=False)
# Customize the z axis
ax.set_zlim(0,2)
ax.zaxis.set_major_locator(LinearLocator(6))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.01f'))
ax.set_xlabel('xaxis')
ax.set_ylabel('yaxis')
ax.set_zlabel('phi(x,y)')
# Add a color bar which maps values to colors
fig.colorbar(surf,shrink=0.75,aspect=10)
plt.show()
```
